# Process the data

```python
import pandas as pd
```

## Study 1

```python
study1 = pd.read_csv("study1_processed.csv")          

# recode sex as a meaningful factor name
study1['sex'] = study1['sex'].map({1: 'female', 2: 'male'})

# calculate BMI
study1["BMI"] = study1["weight_kg"] / (study1["height_meters"]**2)

# calculate waist / height
study1["WHtR"] = study1["waist_meters"] / study1["height_meters"]

# calculate average subjectiuve hunger score across the 15 questions
hunger = study1.loc[:,study1.columns.str.startswith("hunger")]
study1["subjective_hunger"] = hunger.mean(axis=1)
# drop the raw hunger scores
study1 = study1.loc[:,~study1.columns.str.startswith('hunger')]

# drop other columns
study1 = study1.drop(["StartDate", "CompletionDate"], axis=1)
```

```python
study1.head()
```

```python
study1.to_csv("study1_processed.csv", index=False)
```
